```python
# customers: OK
# geolocation: OK
# order_items: shipping_limit_date - change type object-> datetime
# orders: order_purchase_timestamp/ order_approved_at/ order_delivered_carrier_date/ order_delivered_customer_date/ order_estimated_delivery_date
# payments: OK
# products: product_name_length/ product_description_length/ product_photos_qty -> int
# sellers: OK
```

```python
# order_items: shipping_limit_date - change type object-> datetime

order_items['shipping_limit_date'] = pd.to_datetime(
    order_items['shipping_limit_date'].astype(str),
    errors='coerce'
)
order_items.info()
```

```python
# orders: order_purchase_timestamp/ order_approved_at/ order_delivered_carrier_date/ order_delivered_customer_date/ order_estimated_delivery_date

date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

orders[date_cols] = orders[date_cols].apply(
    pd.to_datetime, format='%m/%d/%Y %H:%M', errors='coerce'
)

orders.info()
```

```python
# products: product_name_length/ product_description_length/ product_photos_qty -> int

product_need_int = [
    'product_name_length',
    'product_description_length',
    'product_photos_qty'
]

products[product_need_int] = products[product_need_int].astype(int)

products.info()
```
